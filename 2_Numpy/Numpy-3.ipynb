```python
#Filter values > mean
arr=np.array(list(map(int,input("Enter values of array:").split())))
def val_greater_mean(arr):
    mean_val=np.mean(arr)
    filtered_arr=list(map(int,(filter(lambda x:x>mean_val,arr))))
    return mean_val,filtered_arr
print(val_greater_mean(arr))
```

```python
#Normalize array
import numpy as np
arr=np.array(list(map(int,input("Enter nums:").split())))
def normalize(arr):
    scale=(arr-arr.min())/(arr.max()-arr.min())
    return list(map(float,scale))
print(normalize(arr))
```

```python
#Normalize array
arr=np.array(list(map(int,input("Enter values:").split())))
def standard_scale(arr):
    scale=(arr-arr.mean())/(arr.std())
    return list(map(float,scale))
print(standard_scale(arr))
```

```python
#Normalize a 3×3 matrix
r,c=list(map(int,input("Enter rows and columns:").split()))
def read_matx():
    matx_3=np.array([list(map(int,input(f"Enter {c} values:").split())) for _ in range(r)])
    return matx_3
matx=read_matx()
def normalize(matx):
    scaling=(matx-matx.min())/(matx.max()-matx.min())
    return scaling
print(normalize(matx))
```

```python
#Normalize column-wise
r,c=list(map(int,input("Enter rows and columns:").split()))
def read_matx():
    matx_3=np.array([list(map(int,input(f"Enter {c} values:").split())) for _ in range(r)])
    return matx_3
matx=read_matx()
def normalize(matx):
    scaling=(matx-matx.min(axis=0))/(matx.max(axis=0)-matx.min(axis=0))
    return scaling
print(normalize(matx))
```
