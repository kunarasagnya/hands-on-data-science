```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, cauchy, poisson,expon, gamma,lognorm, binom, weibull_min
```

```python
uniform_random_numbers=np.random.uniform(0,1,1000)
print("random numbers generated from standard uniform distribution with parameters 0,1:", uniform_random_numbers)
plt.subplot(2, 4, 1)
plt.hist(uniform_random_numbers,bins=30,density=True)
plt.title("Uniform Distribution ")
```

```python
random_numbers=np.random.normal(0,1,1000)
print("random numbers generated from standard normal distribution with parameters 0,1:", random_numbers)
plt.subplot(2, 4, 2)
plt.hist(random_numbers,bins=30,density=True)
plt.title("Normal Distribution ")
```
