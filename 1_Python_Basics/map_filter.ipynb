```python
nums=list(map(int,input().split()))
def even(nums):
    return list(filter(lambda x:x%2==0,nums))
def squares(nums):
    return list(map(lambda x:x**2,even(nums)))
print(squares(nums))
```

```python
marks=list(map(int,input().split()))
def add_five(marks):
    return list(map(lambda x:x+5,marks))
print(add_five(marks))
```

```python
marks=list(map(int,input().split()))
def pass_marks(marks):
    return list(filter(lambda x:x>=40,marks))
print(pass_marks(marks))
```

```python
vals=list(map(int,input().split()))
mean=sum(vals)//len(vals)
def sub_mean(vals):
    return list(map(lambda x:x-mean,vals))
print(sub_mean(vals))
```

```python
demand=list(map(int,input().split()))
def avg(demand):
    return sum(demand)/len(demand)
def high_dem(demand):
    return list(filter(lambda x:x>avg(demand),demand))
print(high_dem(demand))
```

```python
prices=list(map(int,input().split()))
def price_adjust(prices):
    return list(map(lambda x:x+(10/100)*x,prices))
print(price_adjust(prices))
```

```python
chars=list(input())
def vowels(chars):
    return list(filter(lambda x:x in ["a","e","i","o","u"],chars))
print(vowels(chars))
```

```python
words=list(input().split())
def len_word(words):
    return list(map(lambda x:len(x),words))
print(len_word(words))
```

```python
vals=list(input().split())
def no_missing(vals):
    return list(filter(lambda x:x!="None",vals))
print(no_missing(vals))
```

```python
nums=list(map(int,input().split()))
def absolute(nums):
    return list(map(lambda x:abs(x),nums))
print(absolute(nums))
```

```python
names=list(input("Enter names:").split())
def start_with_vowel(names):
    return list(filter(lambda x:x[0].lower() in ["a","e","i","o","u"],names))
print(start_with_vowel(names))
```
