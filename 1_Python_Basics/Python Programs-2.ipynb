```python
#Reverse string
def rev_string(s):
  return s[::-1]
print(rev_string("Rasna"))
```

```python
def is_palindrome(s):
  return s==s[::-1]
print(is_palindrome("WOW"))
```

```python
def fibonacci(n):
  a,b=0,1
  for _ in range(n):
    print(a,end=' ')
    a,b=b,a+b
fibonacci(4)
```

```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)
```

```python
def is_prime(n):
  return n>1 and all(n%i!=0 for i in range(2,int(n**0.5)+1))
is_prime(6)
```

```python
def find_max(lst):
  return max(lst)
find_max([8,9,4])
```

```python
def digit_sum(n):
  return sum(int(i) for i in str(n))
digit_sum(157)
```

```python
def is_armstrong(n):
  digits=str(n)
  power=len(digits)
  return sum(int(d)**power for d in digits)==n
is_armstrong(153)
```

```python
def remove_duplicates(lst):
  return list(set(lst))
remove_duplicates([6,9,7,4,6,7])
```

```python
def custom_sort(lst):
  for i in range(len(lst)):
    for j in range(i+1,len(lst)):
      if lst[i]>lst[j]:
        lst[i],lst[j]=lst[j],lst[i]
  return lst
custom_sort([3,5,2,1,6])
```
