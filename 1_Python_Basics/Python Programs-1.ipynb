```python
import math
a=float(input("Enter a a number:"))
b=float(input("Enter a b number:"))
c=float(input("Enter a c number:"))
d=(b**2)-4*a*c
if d>0:
  root1=(-b+math.sqrt(d))/2*a
  root2=(-b-math.sqrt(d))/2*a
  print(f"Root1:{root1}")
  print(f"Root2:{root2}")
elif d==0:
  root=-b/2*a
  print(f"Root:{root}")
else:
  real_part=-b/2*a
  imaginary_part=math.sqrt(abs(d))/2*a
  root1=f"{real_part}+{imaginary_part}*i"
  root2=f"{real_part}-{imaginary_part}*i"
  print(f"Root1:{root1}")
  print(f"Root2:{root2}")
```

```python
year=int(input("Enter a year:"))
if (year%4==0 and year%100==0):
  print("Leap Year")
elif (year%4==0 and year%100!=0):
  print("Leap Year")
else:
  print("Not a Leap Year")
```

```python
num=int(input("Enter a number"))
count=0
for i in range(1,num+1):
  if num%i==0:
    count+=1
if count==2:
  print("Prime Number")
else:
  print("Not a Prime Number")
```
