**Assigning the values for X and y variables **

```python
X=car_data.iloc[:,[1,3,4,6]].values
y=car_data.iloc[:,2].values
```

# **Converting the categorical variables to numerical using LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
X[:,2]=lb.fit_transform(X[:,2])
lb1=LabelEncoder()
X[:,3]=lb.fit_transform(X[:,3])
```

```python
print(X.shape)
```

# **Splitting the data into training and testing**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train[0,:])
```

# **Implementing the RandomForestRegressor Model**

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=400,random_state=0)
regressor.fit(X_train,y_train)
```

# **Accuracy of Model**

```python
accuracy=regressor.score(X_test,y_test)
print(accuracy*100)
```

# **Making Prediction on new data**

```python
new_cardata=[2015,50000,"Petrol","Manual"]
new_cardata[2]=lb.transform([new_cardata[2]])[0]
new_cardata[3]=lb1.transform([new_cardata[3]])[0]
```
