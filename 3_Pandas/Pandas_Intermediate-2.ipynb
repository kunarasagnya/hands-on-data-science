Giving Index and column names

```python
scores=np.array([[9,8,10,7,9],[8,9,9,6,8],[9,10,7,8,9]])
subjects=["Maths","Physics","Chemistry"]
column_names=["Sai Lakshmi","Rasagnya","Akhila","Kalyani","Naga Sai"]
marks_of_students=pd.DataFrame(scores,index=subjects,columns=column_names)
print(marks_of_students)
```
