```python
import pandas as pd
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4,5],"Class":["DS","ML","Python","Java","SQL"]})
df1
```

```python
df2=pd.DataFrame({"ID":[4,5,6,7,8],"Name":["Pramila","Lalli","Samyu","Rasagnya","Shakshitha"]})
df2
```

```python
pd.merge(df1, df2, on="ID")
```

```python
pd.merge(df2, df1)
```

```python
pd.merge(df1, df2, on="ID", how="inner")
```

```python
pd.merge(df1, df2, on="ID", how="left")
```

```python
pd.merge(df1, df2, on="ID", how="right")
```

```python
pd.merge(df1, df2, on="ID", how="outer")
```

```python
pd.merge(df1, df2, on="ID", how="outer",indicator=True)
```

```python
df3=pd.DataFrame({"ID":[6,7,8,9],"age":[23,21,25,19]})
df3
```

```python
pd.merge(df1, df3)
```

```python
pd.merge(df1, df3, left_index=True, right_index=True)
```

```python
df4=pd.DataFrame({"ID":[1,2,3,4,5],"Class":["DS","ML","Python","Java","SQL"]})
df4
```

```python
pd.merge(df1, df4,on="ID",suffixes=["_higher","_middle"])
```
