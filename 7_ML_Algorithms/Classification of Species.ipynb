# **Importing the required Libraries**

```python
import numpy as np
import pandas as pd
```

```python
iris=pd.read_csv("IRIS.csv")
```

```python
X=iris.iloc[:,0:4].values
y=iris.iloc[:,4].values
```

# **Converting the categorical variable to numerical form**

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
y=lb.fit_transform(y)
```

# **Splitting the data into training and testing**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

# **Training the model through Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100,random_state=0)
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

# **Evaluating the model**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(np.argmax(y_pred,axis=1),np.argmax(y_test,axis=1))
accuracy=accuracy_score(np.argmax(y_pred,axis=1),np.argmax(y_test,axis=1))
```

```python
print(f"Confustion Matrix: \n{cm}")
```

```python
print(f"Accuracy Scores is {accuracy*100}")
```
