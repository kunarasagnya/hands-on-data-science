**Handling missing Values by droping and filling Measures of Central Tendency values**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/train.csv")
df.head()
```

```python
if (df["MSSubClass"]>60 and df["LotFrontage"]>60):
  print(df['Id','MSSubClass','LotArea'])
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull())
```

```python
null_vars=df.isnull().sum()/df.shape[0]*100
```

```python
missing_vars=null_vars[null_vars>20].keys()
```

```python
df_drop_clms=df.drop(columns=missing_vars)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_drop_clms.isnull())
```

```python
df_drop_clms=df_drop_clms.select_dtypes(include=["int","float"])
```

```python
missing_num_vars=[var for var in df_drop_clms.columns if df_drop_clms[var].isnull().sum()>0]
print(missing_num_vars)
```

```python
plt.figure(figsize=(8,6))
sns.set()
for i,var in enumerate(missing_num_vars):
  plt.subplot(2,2,i+1)
  sns.histplot(df[var],bins=20,kde=True)
```

```python
df_mean=df_drop_clms.fillna(df_drop_clms.mean())
```

```python
df_median=df_drop_clms.fillna(df_drop_clms.median())
```

```python
plt.figure(figsize=(12,12))
sns.set()
for i,var in enumerate(missing_num_vars):
  plt.subplot(3,1,i+1)
  sns.histplot(df[var],bins=20,kde=True,color="r")
  sns.histplot(df_mean[var],bins=20,kde=True,color="b")
  sns.histplot(df_median[var],bins=20,kde=True,color="g")
```

```python
for i,var in enumerate(missing_num_vars):
  plt.figure(figsize=(12,12))
  plt.subplot(3,1,1)
  sns.boxplot(df[var])
  plt.subplot(3,1,2)
  sns.boxplot(df_mean[var])
  plt.subplot(3,1,3)
  sns.boxplot(df_median[var])
```

```python
df_concat=pd.concat([df[missing_num_vars],df_mean[missing_num_vars],df_median[missing_num_vars]],axis=1)
```
